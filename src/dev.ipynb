```python
import subprocess as sp
from pathlib import Path
from typing import Optional
class DirectoryNavigator:
    def __init__(self,start_dir = '.'):
        self.start_dir = Path(start_dir)
        self.current_dir = self.start_dir
        
    def change_dir(self, new_dir : str, is_relative : bool = False):
        """ Changes directory

        Args:
          new_dir: the new directory to navigate to
          is_relative : (whether the directory is relative) False
        """
        if is_relative:
            self.current_dir = (self.current_dir / new_dir).resolve()
        else:
            self.current_dir = Path(new_dir).resolve()

    def get_dir_tree(self):
        result = sp.run(
            ["tree", self.current_dir], 
            capture_output = True, 
            text = True, 
            check = True
        ) 
        return result.stdout
    
    def run_ls(self, params : Optional[list[str]]  = None):
        """Runs the 'ls' linux command with parameters 
        """
        _call = ['ls']
        if params is not None:
            _call.append(*params)

        result = sp.run(
            _call, 
            capture_output = True,
            text = True, 
            check = True
        )
        return result.stdout
```

```python
d = DirectoryNavigator('.')
```
